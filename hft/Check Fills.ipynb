```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df = pd.read_csv('trades.log', header=None, names=['TYPE', 'TIME', 'TICKER', 'PRICE', 'QUANTITY', 'DIR', 'orderID', 'orderID_mine'])
```

```python
df_buys = df[df['DIR'] == 1]
df_sells = df[df['DIR'] == 0]

plt.figure(figsize=(20,12))
plt.scatter(df_buys['TIME'], df_buys['PRICE'], c='g')
plt.scatter(df_sells['TIME'], df_sells['PRICE'], c='r')
plt.plot(df['TIME'], df['PRICE'])
plt.show()
```
