```python
N = 100

data = np.genfromtxt('history.csv',delimiter=',')
returns = data[1:N,:] - data[:N-1,:]
corr = np.corrcoef(returns.T)

print(len(data))

sns.heatmap(corr, annot=True) ## use the median ---> highly likely to be correct
plt.show()

rho = np.median(corr.reshape(-1))
rho
```

```python
X = returns[:,:2]
Y = returns[:,2:]
Y_pred = X * rho
Y-Y_pred
```
